```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
gpa = pd.read_csv('data/5bee2b6565a8a4015d5e134f.csv')
depression = pd.read_csv('data/5bee365265a8a4015d5f2f2f.csv')
```

```python
gpa
```

```python
combined = gpa.merge(pd.concat([depression['idHash'], depression.iloc[:,14:]], axis=1), on='idHash', how='inner')
```

```python
combined = combined.loc[
    ~combined['How important do you consider your GPA?'].isnull()
]
```

```python
depressed = combined.loc[
    combined['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This applies to me'
]
not_depressed = combined.loc[
    combined['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This does not apply to me'
]
```

```python
depressed['What is your GPA?'].value_counts()
```

```python
GPA_INDICES = [
    '<2.50',
    '2.50-2.59',
    '2.60-2.69',
    '2.70-2.79',
    '2.80-2.89',
    '2.90-2.99',
    '3.00-3.09',
    '3.10-3.19',
    '3.20-3.29',
    '3.30-3.39',
    '3.40-3.49',
    '3.50-3.59',
    '3.60-3.69',
    '3.70-3.79',
    '3.80-3.89',
    '3.90-4.0'
]
```

```python
DEPRESSED_SUM = depressed['What is your GPA?'].value_counts().sum() - depressed['What is your GPA?'].value_counts()['Not applicable']

y_pos = np.arange(len(GPA_INDICES))
plt.figure(figsize=(20, 3))
plt.bar(y_pos, 
        [depressed['What is your GPA?'].value_counts()[index] / DEPRESSED_SUM for index in GPA_INDICES], 
        align='center', 
        alpha=0.5)
plt.xticks(y_pos, GPA_INDICES)
plt.show()
```

```python
NOT_DEPRESSED_SUM = not_depressed['What is your GPA?'].value_counts().sum() - not_depressed['What is your GPA?'].value_counts()['Not applicable']

plt.figure(figsize=(20, 3))
plt.bar(y_pos, 
        [not_depressed['What is your GPA?'].value_counts()[index] / NOT_DEPRESSED_SUM for index in GPA_INDICES], 
        align='center', 
        alpha=0.5)
plt.xticks(y_pos, GPA_INDICES)
plt.show()
```

```python
depression_grades = pd.DataFrame({
    'Grade Bin' : GPA_INDICES,
    'Depressed Grades' : [depressed['What is your GPA?'].value_counts()[index] / DEPRESSED_SUM for index in GPA_INDICES],
    'Not Depressed Grades' : [not_depressed['What is your GPA?'].value_counts()[index] / NOT_DEPRESSED_SUM for index in GPA_INDICES]
})
```
