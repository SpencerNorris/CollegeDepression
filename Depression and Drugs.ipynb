```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
depression = pd.read_csv('data/5bee365265a8a4015d5f2f2f.csv')
drugs = pd.read_csv('data/5beff972ce63f50325d659eb.csv')
data = depression.merge(pd.concat([drugs['idHash'], drugs.iloc[:,14:]], axis=1), on='idHash', how='inner')
```

```python
suicidal = data.loc[
    data['I have experienced suicidal thoughts.'] == 'This applies to me'
]
suicidal
```

```python
depressed = data.loc[
    data['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This applies to me'
]
not_depressed = data.loc[
    data['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This does not apply to me'
]
depressed
```

```python
#We want to get tallies for the drugs that depressed students are taking
unique_drugs = {}
for collection in depressed['Which of the following drugs have you done?']:
    try:
        for drug in collection.split(','):
            if not drug in unique_drugs.keys():
                unique_drugs[drug] = 1
            else:
                unique_drugs[drug] += 1
    except AttributeError:
        print(collection)
```
