```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
depression = pd.read_csv('data/5bee365265a8a4015d5f2f2f.csv')
depression
```

```python
depressed = depression.loc[
    depression['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This applies to me'
]
not_depressed = depression.loc[
    depression['I often experience symptoms of depression (e.g., sadness, anxiety, apathy, loss of interest or pleasure in activities, hopelessness, fatigue, etc.)'] == 'This does not apply to me'
]
```

```python
depressed_by_major = {}
for majors in depressed['major']:
    try:
        for maj in majors.split(','):
            val = maj.strip().lower()
            if not val in depressed_by_major.keys():
                depressed_by_major[val] = 1
            else:
                depressed_by_major[val] += 1
    except AttributeError:
        continue           
# by_major = pd.DataFrame(by_major)
```

```python
not_depressed_by_major = {}
for majors in not_depressed['major']:
    try:
        for maj in majors.split(','):
            val = maj.strip().lower()
            if not val in not_depressed_by_major.keys():
                not_depressed_by_major[val] = 1
            else:
                not_depressed_by_major[val] += 1
    except AttributeError:
        continue  
```

```python
#Get percentages of people by major that are depressed
percentages = {}
for k in depressed_by_major.keys():
    if not k in not_depressed_by_major.keys():
        continue
    else:
        percentages[k] = depressed_by_major[k] / (depressed_by_major[k] + not_depressed_by_major[k])
```

```python
percentages
```

```python
#Let's limit our dataset to majors that had > 100 respondents
keepers = []
for k in depressed_by_major.keys():
    if k in not_depressed_by_major.keys() and (depressed_by_major[k] + not_depressed_by_major[k]) > 100:
        keepers.append(k)
```

```python
len(keepers)
```

```python
keepers
```

```python
percentdf = pd.DataFrame({
    'major' : keepers,
    'percent' : [percentages[k] for k in keepers]
})
```

```python
percentdf.sort_values('percent')
```
