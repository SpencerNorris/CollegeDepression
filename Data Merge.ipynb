```python
drugs = pd.read_csv('data/5beff972ce63f50325d659eb.csv')
fighting = pd.read_csv('data/5beffa9ace63f50325d6a9c0.csv')
gpa = pd.read_csv('data/5bf00afbce63f50325d6af46.csv')
politics = pd.read_csv('data/5bf0129bce63f50325d6b518.csv')
depression = pd.read_csv('data/5bee365265a8a4015d5f2f2f.csv')

DATASETS = [drugs, fighting, gpa, politics, depression]
```

```python
#Chop down each sheet so we only have the ID and relevant questions.
#Exclude 'drugs' from that chopping so we can keep the demographic data.
for x in range(1,5):
    dataset = DATASETS[x]
    dataset = pd.concat([dataset[['idHash']], dataset.iloc[:,14:]], axis=1)
    DATASETS[x] = dataset
data = reduce(lambda x,y: x.merge(y, on='idHash', how='inner'), DATASETS)
data
```
